**Importing Pandas**

```python
import pandas as pd
```

**Load Data**

```python
customers = pd.read_csv('https://raw.githubusercontent.com/karthikeyan476/Data-Analysis-/refs/heads/main/customers%20(1).csv')  #Read command is used to read the file
transactions = pd.read_csv('https://raw.githubusercontent.com/karthikeyan476/Data-Analysis-/refs/heads/main/transactions.csv')
campaigns = pd.read_csv('https://raw.githubusercontent.com/karthikeyan476/Data-Analysis-/refs/heads/main/campaigns.csv')
```

**Joining Tables**

```python
df_merged = pd.merge(transactions, customers, on='customer_id', how='left')   #Merge transactions with customers on 'customer_id'
```

```python
df_final = pd.merge(df_merged, campaigns, on='customer_id', how='left') #Merge the result with campaigns on 'customer_id'
```

**Data Cleansing**

```python
df_final['gender'] = df_final['gender'].str.lower().str.strip().replace({'m': 'male', 'f': 'female'})
```

```python
print (df_final)
```

```python
df_final = df_final.drop_duplicates()
```

```python
df_final['platform'] = df_final['platform'].fillna('Unknown')
df_final['impressions'] = df_final['impressions'].fillna(0).astype(int)
df_final['clicks'] = df_final['clicks'].fillna(0).astype(int)
df_final['conversions'] = df_final['conversions'].fillna(0).astype(int)
df_final['spend'] = df_final['spend'].fillna(0.0)
```

**Data Wrangling**

```python
df_final['transaction_date'] = pd.to_datetime(df_final['transaction_date'])         #Extract date parts
df_final['transaction_month'] = df_final['transaction_date'].dt.month
```

```python
print(df_final)
```

```python
df_final['total_spend'] = df_final['amount'] + df_final['spend']
```

```python
print (df_final = 'total_spend')
```

```python
df_final['ROAS'] = df_final['amount'] / df_final['spend']
df_final['ROAS'] = df_final['ROAS'].replace([float('inf'), -float('inf')], 0).fillna(0)
```

```python
df_final['age_group'] = pd.cut(df_final['age'], bins=[0, 25, 35, 45, 55, 100],
                               labels=['<25', '25-35', '35-45', '45-55', '55+'])
```

```python
print (df_final['age_group'])
```
